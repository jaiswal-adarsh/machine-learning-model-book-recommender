```python
import nltk
```

```python
#stemming to extraxt the root only

from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
#defining stemming function

def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
new_df['tags']= new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
#convering to lower case 
new_df['tags']= new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.tail()
```

```python
#Need to vectorize the tags for making comparisions for recommendation
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000, stop_words= 'english')
```

```python
vectors= cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
 from sklearn.metrics.pairwise import cosine_similarity
```

```python
#finding the cosine similarity between the vectors
similarity= cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
#creating function to generate similar 5 movies from vector
def recommend (book):
    book_index= new_df[new_df['b_title'] == book].index[0]
    distances= similarity[book_index]
    book_list= sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:21]
    
    for i in book_list:
        print(new_df.iloc[i[0]].b_title)
    return
```

```python
recommend('Santa Paws')
```
